```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean1=55
std_dv1=10
num_samples=500

column1_numbers=np.random.normal(mean1,std_dv1,num_samples)
column1_numbers=np.clip(column1_numbers,30,120)
column1_numbers=np.round(column1_numbers).astype(int)

mean2=18
std_dv2=5

column2_numbers=np.random.normal(mean2,std_dv2,num_samples)
column2_numbers=np.clip(column2_numbers,12,26)
column2_numbers=np.round(column2_numbers).astype(int)

column3_numbers=np.random.randint(2,size=num_samples)
column3_numbers[column3_numbers>mean1]=1

data={
    'Miles Per_week':column1_numbers,
    'Farthest_run':column2_numbers,
    'Qualified_Boston_Marathon':column3_numbers
}
df=pd.DataFrame(data)
```

```python
df
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Miles Per_week'],df['Farthest_run'],c=df['Qualified_Boston_Marathon'],cmap='coolwarm')
plt.xlabel('Miles Per_week')
plt.ylabel('Farthest_run')
plt.title('Scatter Plot of Miles Per_week vs Farthest_run')
plt.colorbar(label='Qualified_Boston_Marathon')
plt.show()
```

```python
X=df.iloc[:,:2]
```

```python
X
```

```python
y=df.iloc[:,2]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=33,test_size=0.2)
```

```python
from sklearn.svm import SVC
```

```python
model=SVC()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

#Regularization

```python
model_reg0=SVC(C=0.1)
```

```python
model_reg0.fit(X_train,y_train)
```

```python
model_reg0.score(X_test,y_test)
```

```python
model_reg1=SVC(C=1)
model_reg1.fit(X_train,y_train)
model_reg1.score(X_test,y_test)
```

```python
model_reg2=SVC(C=1000)
model_reg2.fit(X_train,y_train)
model_reg2.score(X_test,y_test)
```

#gamma

```python
model_gamma0=SVC(gamma=0.1)
model_gamma0.fit(X_train,y_train)
model_gamma0.score(X_test,y_test)
```

```python
model_gamma1=SVC(gamma=1)
model_gamma1.fit(X_train,y_train)
model_gamma1.score(X_test,y_test)
```

```python
model_gamma2=SVC(gamma=1000)
model_gamma2.fit(X_train,y_train)
model_gamma2.score(X_test,y_test)
```

#kerneL

```python
model_liner=SVC(kernel='linear')
model_liner.fit(X_train,y_train)
model_liner.score(X_test,y_test)
```

```python
model_liner1=SVC(kernel='rbf')
model_liner1.fit(X_train,y_train)
model_liner1.score(X_test,y_test)
```
