```python
import pandas as pd
import numpy as np
```

```python
mean1=55
std_dv1=10
num_samples=500

column1_numbers=np.random.normal(mean1,std_dv1,num_samples)
column1_numbers=np.clip(column1_numbers,30,120)
column1_numbers=np.round(column1_numbers).astype(int)

mean2=18
std_dv2=3

column2_numbers=np.random.normal(mean2,std_dv2,num_samples)
column2_numbers=np.clip(column2_numbers,12,26)
column2_numbers=np.round(column2_numbers).astype(int)

column3_numbers=np.random.randint(2,size=num_samples)
column3_numbers[column3_numbers>mean1]=1

data={
    'Miles Per_week':column1_numbers,
    'Farthest_run':column2_numbers,
    'Qualified_Boston_Marathon':column3_numbers
}
df=pd.DataFrame(data)
```

```python
df.head(10)
```

```python
X=df.iloc[:,:2]
```

```python
y=df.iloc[:,2]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=26)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
```

```python
param_grid = [{
    'n_estimators': [500, 1000, 1500],
    # 'criterion': ['gini', 'entropy'],
    'min_samples_split': [5, 10, 15],
    'min_samples_leaf': [1, 2, 4]  # Corrected spelling
    # 'max_depth': [10, 20, 30]
}]
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_search = GridSearchCV(rf, param_grid, scoring='accuracy', cv=2, n_jobs=-1)
```

```python
grid_search.fit(X_train, y_train)
```

```python
grid_search.best_score_
```

```python
grid_search.best_params_
```

#randomized

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
random_param_grid = [{
    'n_estimators': [500, 1000, 1500],
    'criterion': ['gini', 'entropy'],
    'min_samples_split': [5, 10, 15],
    'min_samples_leaf': [1, 2, 4],
    'max_depth': [10, 20, 30]
}]
```

```python
random_grid_search = RandomizedSearchCV(rf, random_param_grid, scoring='accuracy', cv=5, n_jobs=-1,random_state=26)
```

```python
random_grid_search.fit(X_train, y_train)
```

```python
random_grid_search.best_score_
```

```python
random_grid_search.best_params_
```
