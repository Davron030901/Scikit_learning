```python
import pandas as pd
```

```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
mpg=sns.load_dataset('mpg')
```

```python
mpg.head()
```

```python
mpg=mpg.drop('name',axis=1)
```

```python
mpg=pd.get_dummies(mpg,dtype=int)
```

```python
mpg.head()
```

```python
pd.DataFrame(mpg.isnull().sum().sort_values(ascending=False))
```

```python
mpg['horsepower'].fillna(mpg['horsepower'].mean(),inplace=True)
```

```python
pd.DataFrame(mpg.isnull().sum().sort_values(ascending=False))
```

```python
X=mpg.drop('mpg',axis=1)
y=mpg['mpg']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=19)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mean_squared_error(y_test,y_pred)
```

```python
r2_score(y_test,y_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rfr=RandomForestRegressor(random_state=13)
```

```python
rfr.fit(X_train,y_train)
```

```python
y_pred2=rfr.predict(X_test)
```

```python
mean_squared_error(y_test,y_pred2)
```

```python
r2_score(y_test,y_pred2)
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge=Ridge()
```

```python
param_grid={'alpha':[0.05,0.1,0.3,1,3,5,10,15,30,50,75]}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
ridge_cv=GridSearchCV(ridge,param_grid,cv=5,n_jobs=-1,scoring='neg_mean_squared_error')
```

```python
ridge_cv.fit(X_train,y_train)
```

```python
ridge_cv.best_estimator_
```

```python
y_pred3=ridge_cv.predict(X_test)
```

```python
mean_squared_error(y_test,y_pred3)
```

```python
r2_score(y_test,y_pred3)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
gbr=GradientBoostingRegressor()
```

```python
gbr.fit(X_train,y_train)
```

```python
y_pred4=gbr.predict(X_test)
```

```python
mean_squared_error(y_test,y_pred4)
```

```python
r2_score(y_test,y_pred4)
```

```python
from sklearn.ensemble import StackingRegressor
```

```python
estemators=[
    ('lr',lr),
    ('ridge',ridge_cv.best_estimator_),
    ('gbr',gbr),
]
```

```python
sr=StackingRegressor(estimators=estemators,final_estimator=rfr)
```

```python
sr.fit(X_train,y_train)
```

```python
y_pred5=sr.predict(X_test)
```

```python
mean_squared_error(y_test,y_pred5)
```

```python
r2_score(y_test,y_pred5)
```

```python
from sklearn.ensemble import VotingRegressor
```

```python
vr=VotingRegressor(estimators=[
    ('rfr',rfr),
    ('lr',lr),
    ('gbr',gbr),
],weights=[2,3,1])
```

```python
vr.fit(X_train,y_train)
```

```python
y_pred6=vr.predict(X_test)
```

```python
mean_squared_error(y_test,y_pred6)
```

```python
r2_score(y_test,y_pred6)
```

```python
sr2=StackingRegressor(estimators=estemators,final_estimator=vr)
```

```python
sr2.fit(X_train,y_train)
```

```python
y_pred7=sr2.predict(X_test)
```

```python
mean_squared_error(y_test,y_pred7)
```

```python
r2_score(y_test,y_pred7)
```

```python
from sklearn.svm import SVR
```

```python
estemators=[
    ('ridge',ridge_cv.best_estimator_),
    ('gbr',gbr),
    ('svr',SVR(C=1.0,kernel='linear')),
    ('random_forest',RandomForestRegressor())
]
```

```python
sr3=StackingRegressor(estimators=estemators,final_estimator=Ridge(alpha=1.0))
```

```python
param_grid={
    'random_forest__n_estimators':[50,100,250],
    'svr__C':[0.1,1.0,10.0],
    'final_estimator__alpha':[0.1,1.0,10.0],
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
sr_cv=RandomizedSearchCV(sr3,param_grid,n_iter=5,cv=3,n_jobs=-1,scoring='neg_mean_squared_error')
```

```python
sr_cv.fit(X_train,y_train)
```

```python
y_pred8=sr_cv.predict(X_test)
```

```python
mean_squared_error(y_test,y_pred8)
```

```python
r2_score(y_test,y_pred8)
```

```python
sr_cv.best_estimator_
```
