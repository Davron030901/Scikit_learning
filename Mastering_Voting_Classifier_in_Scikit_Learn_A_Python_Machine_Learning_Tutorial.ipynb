```python
import pandas as pd
```

```python
from sklearn.datasets import make_classification
```

```python
X,y=make_classification(n_samples=2000,n_features=10,n_informative=8,n_redundant=2,random_state=11)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=11)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
gnb=GaussianNB()
```

```python
gnb.fit(X_train,y_train)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cross_val_score(gnb,X_train,y_train,cv=3).mean()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
cross_val_score(lr,X_train,y_train,cv=3).mean()
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc=RandomForestClassifier()
```

```python
rfc.fit(X_train,y_train)
```

```python
cross_val_score(rfc,X_train,y_train,cv=3).mean()
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
vc=VotingClassifier(estimators=[('NaiveBayes',gnb),('LogisticRegression',lr),('RandomForest',rfc)])
```

```python
cross_val_score(vc,X_train,y_train,cv=3).mean()
```

```python
param_grid={
    'voting':['hard','soft'],
    'weights':[(1,1,1),(2,1,1),(1,2,1),(1,1,2)]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
vc2=GridSearchCV(vc,param_grid=param_grid,cv=5,n_jobs=-1)
```

```python
vc2.fit(X_train,y_train)
```

```python
vc2.best_params_
```

```python
vc2.best_score_
```
